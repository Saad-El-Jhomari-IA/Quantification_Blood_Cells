```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
import random

extracted_dir = Path("../data/processed/extracted_cells")

if not extracted_dir.exists():
    print("[WARNING] Extracted cells directory not found. Run scripts/05_extract_wbc_patches.py first.")
else:
    patch_files = list(extracted_dir.glob("*.jpg"))
    print(f"[INFO] Total WBC patches extracted: {len(patch_files)}")

    if patch_files:
        sample = random.sample(patch_files, min(76, len(patch_files)))
        fig, axes = plt.subplots(7, 7, figsize=(10, 10))

        for i, ax in enumerate(axes.flat):
            if i < len(sample):
                img = cv2.imread(str(sample[i]))
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                ax.imshow(img)
                ax.set_title(f"#{i+1}", fontsize=8)
            ax.axis('off')

        plt.suptitle(f"Extracted WBC Patches (Validation Set) - Sample of {len(sample)}", fontsize=14)
        plt.tight_layout()
        plt.show()
    else:
        print("[INFO] No patches found. The validation set might not contain WBCs or the model detected none.")
```
